```python
# --- Cell 1: Imports (Corrected) ---
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import optimizers, applications
# CORRECTED PATH
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# Use ResNet50 or the appropriate model for SHAP/LIME
from tensorflow.keras.applications.resnet50 import preprocess_input
import cv2
import shap # NEW: For explainability
from fpdf import FPDF # NEW: For PDF generation

# --- Cell 2: Drive Mount (New) ---
from google.colab import drive
drive.mount('/content/drive')
BASE_DIR = '/content/drive/MyDrive/Explainability Exercise/'

# --- Cell 3: Load Data (Path Check) ---
# Load data
train = pd.read_csv('/content/drive/MyDrive/Explainability Exercise/data/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Explainability Exercise/data/test.csv')
print('Number of train samples: ', train.shape[0])
print('Number of test samples: ', test.shape[0])

# Preprocecss data
train["id_code"] = train["id_code"].apply(lambda x: x + ".png")
test["id_code"] = test["id_code"].apply(lambda x: x + ".png")
train['diagnosis'] = train['diagnosis'].astype('str')
test['diagnosis'] = test['diagnosis'].astype('str')
display(train.head())
display(test.head())

# --- Cell 4 (New): Model Loading ---
model = tf.keras.models.load_model(
    f'{BASE_DIR}model.h5',
    compile=False
)
print("Model loaded successfully for inference.")
```

Task A: Patient Selection and Preprocessing Logic
This code finds a specific High-Risk and Low-Risk patient from your dataframe and defines a function to prepare their images for the model (matching the training scaling of 1./255).

```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# 1. Configuration & Data Cleaning
# FIX: Set the image size to 320x320 to match the model's required input shape
IMG_SIZE = (320, 320)
IMG_DIR_TEST = f'/content/drive/MyDrive/Explainability Exercise/data/test_images/'

try:
    test['diagnosis'] = test['diagnosis'].astype(int)
    train['diagnosis'] = train['diagnosis'].astype(int)
except Exception as e:
    print(f"Warning: Could not convert diagnosis column to int. Error: {e}")

test['id_code'] = test['id_code'].str.replace('.png', '', regex=False)
train['id_code'] = train['id_code'].str.replace('.png', '', regex=False)
print("ID codes cleaned of '.png' extension.")

# 2. Image Preprocessing Function
def get_processed_image(id_code, base_image_dir):
    """
    Loads an image by ID, resizes it to IMG_SIZE, normalizes it (0-1),
    and adds a batch dimension.
    """
    file_path = f"{base_image_dir}{id_code}.png"

    # Load and Resize
    original_img = load_img(file_path, target_size=IMG_SIZE)

    # Convert to Array and Normalize (0 to 1)
    img_array = img_to_array(original_img)
    img_array = img_array / 255.0

    # Add Batch Dimension (1, 320, 320, 3)
    img_tensor = np.expand_dims(img_array, axis=0)

    return img_tensor, original_img

# 3. Patient Selection Logic
print("\nSelecting specific patients from the test set...")
high_risk_rows = test[test['diagnosis'] >= 3] # Class 3 or 4
low_risk_rows = test[test['diagnosis'] == 0]  # Class 0

if high_risk_rows.empty or low_risk_rows.empty:
    raise ValueError("Cannot find required patient types in the test data.")

# Pick the first available ID from each group
high_risk_id = high_risk_rows.iloc[0]['id_code']
low_risk_id = low_risk_rows.iloc[0]['id_code']

print(f" High Risk Patient ID: {high_risk_id} (Label: {high_risk_rows.iloc[0]['diagnosis']})")
print(f" Low Risk Patient ID:  {low_risk_id} (Label: {low_risk_rows.iloc[0]['diagnosis']})")

# 4. Load and Store Patient Data
X_high, img_high = get_processed_image(high_risk_id, IMG_DIR_TEST)
X_low, img_low = get_processed_image(low_risk_id, IMG_DIR_TEST)
```

Task B: Explanation Generation (SHAP)
This code initializes the SHAP explainer and generates the visualizations. Note: SHAP can be slow. Calculating these values might take 1-2 minutes.

```python
# Run this to fix the Read-only file system error
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
print("\nDrive forcefully remounted. File saving should now work.")
```

```python
# =========================================================================
# TASK B: SHAP EXPLANATION GENERATION (FIXED)
# =========================================================================

import shap
import matplotlib.pyplot as plt

# --- Configuration ---
IMG_DIR_TRAIN = f'{BASE_DIR}data/train_images/'
NUM_BACKGROUND_SAMPLES = 10

# 1. SHAP Background Preparation
print(f"\nPreparing SHAP background ({NUM_BACKGROUND_SAMPLES} images)...")
background_ids = train.sample(NUM_BACKGROUND_SAMPLES)['id_code'].values
background_images = []

for bid in background_ids:
    try:
        # Reuses get_processed_image from Task A
        img_tensor, _ = get_processed_image(bid, IMG_DIR_TRAIN)
        background_images.append(img_tensor[0])
    except FileNotFoundError:
        continue

X_background = np.array(background_images)
print(f"SHAP background prepared with shape: {X_background.shape}")

# 2. Initialize Explainer
print("Initializing SHAP DeepExplainer...")
explainer = shap.DeepExplainer(model, X_background)
print("Explainer initialized. Starting calculation...")


# 3. SHAP Generation and Saving Function
def generate_and_save_shap(img_tensor, patient_id, risk_type):
    """Calculates SHAP values, makes a prediction, and saves the plot to Drive."""
    print(f"\n[DEBUG] Calculating SHAP for {risk_type} patient ID: {patient_id}")

    # Check if explainer is initialized (prevent runtime error if it failed initialization)
    if 'explainer' not in globals():
        print("ERROR: Explainer not initialized. Cannot continue.")
        return 0

    try:
        # Calculate SHAP values
        shap_values = explainer.shap_values(img_tensor)

        # Get Model Prediction
        preds = model.predict(img_tensor)
        predicted_class = np.argmax(preds[0])

        print(f"Model predicted Class: {predicted_class}")

        # FIX: Robust Plotting to avoid "index out of bounds" error.
        # We pass the full shap_values object and tell the function which class to plot
        # using the 'labels' argument. This handles the inconsistent output structure.
        plt.figure(figsize=(6, 6))
        shap.image_plot(
            shap_values,
            img_tensor[0:1],
            labels=[predicted_class], # Tell SHAP to only visualize the predicted class
            show=False
        )

        # Save the figure to Google Drive
        save_path = f"{BASE_DIR}explanation_{risk_type}.png"
        plt.savefig(save_path, bbox_inches='tight')
        plt.close() # Close plot object to clean up memory
        print(f"Saved explanation image to: {save_path}")

        return predicted_class

    except Exception as e:
        print(f"ERROR during SHAP calculation for {risk_type}: {e}")
        return 0 # Return 0 on failure

# 4. Final Execution
p_high_pred = generate_and_save_shap(X_high, high_risk_id, "High_Risk")
p_low_pred = generate_and_save_shap(X_low, low_risk_id, "Low_Risk")
```
